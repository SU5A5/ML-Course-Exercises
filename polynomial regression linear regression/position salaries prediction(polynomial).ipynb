```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Position_Salaries.csv')
X=dataset.iloc[:,1:-1].values
y=dataset.iloc[:,-1].values
print(X)
print(y.reshape(len(y),1))
```

```python
#training the regression model  
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
#trainig the polynomial regression model
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly,y)
```

```python
#visualizing the linear regresssion mmodel
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg.predict(X),color='blue')
plt.title('Position Level vs Salary')
plt.xlabel('Positions Level')
plt.ylabel('Salary')
plt.show()
```

```python
#visualzing the polynomial model
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg2.predict(X_poly),color='blue')
plt.title('Position Level vs Salary(polynomial regression)')
plt.xlabel('Positions Level')
plt.ylabel('Salary')
plt.show()
```

```python
#prediciting the salary using linear regression 
lin_reg.predict([[6.5]])
```

```python
#prediciting the salary using polynomial regression 
lin_reg2.predict(poly_reg.fit_transform([[6.5]]))
```
